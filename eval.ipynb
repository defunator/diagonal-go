```python
for i in range(0, len(df_dict), 2):
    df_dict[i]["speedup"] = df_dict[i+1]["time"] / df_dict[i]["time"]
    df_dict[i]["f_count_overhead"] = (df_dict[i]["FCount"] - df_dict[i+1]["FCount"]) / df_dict[i+1]["FCount"] * 100
    
```

```python
df = pd.DataFrame.from_dict(df_dict).dropna().set_index("name")
```

```python
import matplotlib.pyplot as plt

_, ax = plt.subplots(figsize=(12,5))
plt.bar(df.index, df["speedup"])
plt.xticks(rotation=45, ha='right')
plt.axline((0, 1), (9, 1), color="red")
plt.ylim([0,10])
ax.set_title("Ускорение многопоточной реализации по времени")
plt.grid()
```

```python
import matplotlib.pyplot as plt

_, ax = plt.subplots(figsize=(12,5))
plt.bar(df.index, df["f_count_overhead"])
plt.xticks(rotation=45, ha='right')
ax.set_title("Увеличение числа вызовов $f(x)$ многопоточной реализации в %")
plt.grid()
```
